<span style="font-size:24px; font-weight:bold; color:black;">Create a function for Model Visualization</span>

```python
#initialize result tracking structures
results_list = []  #for storing metrics
fitted_models = {} #for storing trained model objects

#function to evaluate and visualize model performance
def evaluate_model(name, model, X_test, y_test, y_pred, y_prob):
    from sklearn.metrics import classification_report, confusion_matrix, RocCurveDisplay
    import matplotlib.pyplot as plt
    import seaborn as sns
    import numpy as np

    print(f"--- {name} ---\n")
    
    # Classification report
    print("Classification Report:")
    print(classification_report(y_test, y_pred))
    
    # Confusion matrix
    cm = confusion_matrix(y_test, y_pred)
    labels = np.array([["TN", "FP"], ["FN", "TP"]])
    box_labels = np.array([[f"{labels[i, j]}\n{cm[i, j]}" for j in range(2)] for i in range(2)])

    plt.figure(figsize=(5, 4))
    sns.heatmap(cm, annot=box_labels, fmt='', cmap='Blues', cbar=False,
                xticklabels=['Predicted: Not Fraud', 'Predicted: Fraud'],
                yticklabels=['Actual: Not Fraud', 'Actual: Fraud'])
    plt.title(f"{name} - Confusion Matrix")
    plt.xlabel("Prediction")
    plt.ylabel("Actual")
    plt.tight_layout()
    plt.show()

    # ROC curve
    RocCurveDisplay.from_predictions(y_test, y_prob)
    plt.title(f"{name} - ROC Curve")
    plt.grid(True)
    plt.tight_layout()
    plt.show()
#Interpreting the Classification Report:
#precision/fpr: of the claims predicted as fraud, how many were actually fraud?
#recall/tpr: of the actual frauds, how many did the model catch?
#F1-score: a balance of precision and recall (for when you care about both)
#support: how  many true examples existed in each class? (0 = not fraud, 1 = fraud)
#should care a lot about recall for fraud detection, because missing a fraud case can be costly.
```

<span style="font-size:24px; font-weight:bold; color:black;">Prep Data for Logistic Regression</span>

```python
#drop unwanted columns (redundant cuz already encoded or not predictive like identifiers)
columns_to_drop = [
       'Month', 'DayOfWeek', 'Make', 'AccidentArea',
       'DayOfWeekClaimed', 'MonthClaimed', 'WeekOfMonthClaimed', 'Sex',
       'MaritalStatus', 'Fault', 'PolicyType', 'VehicleCategory', 'PolicyNumber',
       'VehiclePrice',  'Days_Policy_Accident','Days_Policy_Claim', 'RepNumber',
       'PastNumberOfClaims', 'AgeOfVehicle', 'AgeOfPolicyHolder', 
       'PoliceReportFiled', 'WitnessPresent', 'AgentType', 'NumberOfSuppliments', 
       'AddressChange_Claim', 'NumberOfCars', 'Year','BasePolicy']

data_model = data.drop(columns = columns_to_drop)
```

```python
#split into input features (X) and target (y)
X = data_model.drop(columns = ['FraudFound_P']) #features
y = data_model['FraudFound_P'] #target
```

```python
#scale the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X) 
```

```python
#train/test split 
#test size of 0.2 = 80% for training, 20% for testing
#using stratified split to preserve class imbalance, ensures same fraud/non-fraud ratio in both sets
#set random state for reproducibility
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size = 0.2, random_state = 42, stratify = y)

print("Data:", Counter(y_train))
```

<span style="font-size:24px; font-weight:bold; color:black;">Logistic Regresssion</span>

```python
#initialize and train the Logistic Regression Model
#class_weight = 'balanced' tells the model to penalize mistakes on minority class, 
#this helps address the imbalance without resampling
lr_model = LogisticRegression(max_iter = 1000, class_weight = 'balanced', random_state = 42)

#fit the model to the training data
#X_train is the features we use to teach the model
#y_train is the correct answers for those features
lr_model.fit(X_train, y_train)

#make predictions
lr_pred = lr_model.predict(X_test) #predicted classes (0 or 1)
lr_proba = lr_model.predict_proba(X_test)[:, 1] #predicted probabilities for class 1 (fraud), the predicted probability that each case is fraud

# Store results
results_list.append({
    "model": "Logistic Regression",
    "precision": precision_score(y_test, lr_pred),
    "recall": recall_score(y_test, lr_pred),
    "f1": f1_score(y_test, lr_pred),
    "roc_auc": roc_auc_score(y_test, lr_proba)
})

# Store trained model
fitted_models["logistic_regression"] = lr_model

# Evaluate visually
evaluate_model("Logistic Regression", lr_model, X_test, y_test, lr_pred, lr_proba)
```

<span style="font-size:24px; font-weight:bold; color:black;">Logistic Regression with SMOTE</span>

```python
# apply SMOTE with k_neighbors=3
smote = SMOTE(k_neighbors=3, random_state=42)
X_train_smote, y_train_smote = smote.fit_resample(X_train, y_train)
print("After SMOTE:", Counter(y_train_smote))

#initialize a new logistic regression model (no need for class_weight here since data is balanced)
lr_model_smote = LogisticRegression(max_iter = 1000, random_state = 42)

#train model on the SMOTE-balanced data
lr_model_smote.fit(X_train_smote, y_train_smote)

#make predictions using the model trained on SMOTE-balanced data
lr_pred_smote = lr_model_smote.predict(X_test) #predicted classes using SMOTE model
lr_proba_smote = lr_model_smote.predict_proba(X_test)[:, 1] #predicted fraud probabilities using SMOTE model

# Store evaluation metrics
results_list.append({
    "model": "Logistic Regression (SMOTE)",
    "precision": precision_score(y_test, lr_pred_smote),
    "recall": recall_score(y_test, lr_pred_smote),
    "f1": f1_score(y_test, lr_pred_smote),
    "roc_auc": roc_auc_score(y_test, lr_proba_smote)
})

# Store trained model
fitted_models["logistic_regression_smote"] = lr_model_smote

# Evaluate visually
evaluate_model("Logistic Regression (SMOTE)", lr_model_smote, X_test, y_test, lr_pred_smote, lr_proba_smote)
```

<span style="font-size:24px; font-weight:bold; color:black;">Compare both Logistic Regression Models</span>

```python
#plot the ROC Curve
#shows tradeoff between true positive rate (recall) and false positive rate (precision)

#calculate FPR and TPR for original model
fpr, tpr, _ = roc_curve(y_test, lr_proba)

#calculate FPR and TPR for SMOTE model
fpr_smote, tpr_smote, _ = roc_curve(y_test, lr_proba_smote)

plt.figure(figsize=(8, 5))

#plot original model ROC curve
plt.plot(fpr, tpr, label=f"Original (AUC = {roc_auc_score(y_test, lr_proba):.2f})")

#plot SMOTE model ROC curve
plt.plot(fpr_smote, tpr_smote, label=f"SMOTE (AUC = {roc_auc_score(y_test, lr_proba_smote):.2f})", color='red')

#plot baseline (random guessing)
plt.plot([0, 1], [0, 1], 'k--') #diagonal line means no skill (random guess)

plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.title("ROC Curve - Logistic Regression vs SMOTE")
plt.legend()
plt.grid()
plt.show()

#AUC = area under the curve, higher is better (max is 1.0, random guess is 0.5)
```

```python
# I used SMOTE to help address class imbalance, as fraud represents about 6% of the data. 
# SMOTE slightly improved precision (from 0.13 to 0.17), reducing false positives. 
# However, it also significantly decreased recall (from 0.90 to 0.29), meaning it missed many actual fraud cases. 
# Since recall is more critical for fraud detection, I chose to stick with the original logistic regression, 
# which identified more fraudulent claims despite its lower precision.
```

```python
#train XGBoost model 
xgb_model = XGBClassifier(use_label_encoder = False, eval_metric = 'logloss', random_state = 42)
xgb_model.fit(X_train2, y_train2)

y_pred_xgb = xgb_model.predict(X_test2)
y_proba_xgb = xgb_model.predict_proba(X_test2)[:, 1]

results_list.append({
    "model": "XGBoost",
    "precision": precision_score(y_test2, y_pred_xgb),
    "recall": recall_score(y_test2, y_pred_xgb),
    "f1": f1_score(y_test2, y_pred_xgb),
    "roc_auc": roc_auc_score(y_test2, y_proba_xgb)
                             })
fitted_models["xgb"] = xgb_model

#train XGBoost model with smote
sm = SMOTE(random_state = 42)
X_train_sm, y_train_sm = sm.fit_resample(X_train2, y_train2)

xgb_sm_model = XGBClassifier(use_label_encoder=False, eval_metric = 'logloss', random_state = 42)
xgb_sm_model.fit(X_train_sm, y_train_sm)

y_pred_xgb_sm = xgb_sm_model.predict(X_test2)
y_proba_xgb_sm = xgb_sm_model.predict_proba(X_test2)[:, 1]

results_list.append({
    "model": "XGBoost (SMOTE)",
    "precision": precision_score(y_test2, y_pred_xgb_sm),
    "recall": recall_score(y_test2, y_pred_xgb_sm),
    "f1": f1_score(y_test2, y_pred_xgb_sm),
    "roc_auc": roc_auc_score(y_test2, y_proba_xgb_sm)
})
fitted_models["xgb_smote"] = xgb_sm_model

#XGBoost with RFE (Recursive Feature Elimination)
rfe_selector = RFE(estimator = XGBClassifier(use_label_encoder = False, eval_metric = 'logloss'),
                   n_features_to_select = 20, step = 1) 
X_train_rfe = rfe_selector.fit_transform(X_train2, y_train2) 
X_test_rfe = rfe_selector.transform(X_test2)

xgb_rfe_model = XGBClassifier(use_label_encoder = False, eval_metric = 'logloss', random_state = 42)
xgb_rfe_model.fit(X_train_rfe, y_train2)

y_pred_xgb_rfe = xgb_rfe_model.predict(X_test_rfe)
y_proba_xgb_rfe = xgb_rfe_model.predict_proba(X_test_rfe)[:, 1]

results_list.append({
    "model": "XGBoost (RFE)",
    "precision": precision_score(y_test2, y_pred_xgb_rfe),
    "recall": recall_score(y_test2, y_pred_xgb_rfe),
    "f1": f1_score(y_test2, y_pred_xgb_rfe),
    "roc_auc": roc_auc_score(y_test2, y_proba_xgb_rfe)
})
fitted_models["xgb_rfe"] = xgb_rfe_model
```

<span style="font-size:24px; font-weight:bold; color:black;">XGBoost with RFE and SMOTE</span>

```python
#apply RFE to the training data
rfe_selector = RFE(estimator = XGBClassifier(use_label_encoder = False, eval_metric = 'logloss'),
                   n_features_to_select = 20, step = 1)
X_train_rfe = rfe_selector.fit_transform(X_train2, y_train2)
X_test_rfe = rfe_selector.transform(X_test2)

#apply SMOTE to RFE-selected features
sm = SMOTE(random_state = 42)
X_train_rfe_sm, y_train_rfe_sm, = sm.fit_resample(X_train_rfe, y_train2)

#train model
xgb_rfe_sm_model = XGBClassifier(use_label_encoder = False, eval_metric = 'logloss', random_state = 42)
xgb_rfe_sm_model.fit(X_train_rfe_sm, y_train_rfe_sm)

y_pred_xgb_rfe_sm = xgb_rfe_sm_model.predict(X_test_rfe)
y_proba_xgb_rfe_sm = xgb_rfe_sm_model.predict_proba(X_test_rfe)[:, 1]   

# threshold = 0.59
# y_pred_xgb_rfe_sm = (y_proba_xgb_rfe_sm >= threshold).astype(int)
# y_proba_xgb_rfe_sm = xgb_rfe_sm_model.predict_proba(X_test_rfe)[:, 1]

#store metrics
results_list.append({
    "model": "XGBoost (RFE + SMOTE)",
    "precision": precision_score(y_test2, y_pred_xgb_rfe_sm),
    "recall": recall_score(y_test2, y_pred_xgb_rfe_sm),
    "f1": f1_score(y_test2, y_pred_xgb_rfe_sm),
    "roc_auc": roc_auc_score(y_test2, y_proba_xgb_rfe_sm)
})
fitted_models["xgb_rfe_smote"] = xgb_rfe_sm_model
```

```python
# get precision, recall, and thresholds
precisions, recalls, thresholds = precision_recall_curve(y_test2, y_proba_xgb_rfe_sm)

# plot
plt.figure(figsize=(8,6))
plt.plot(thresholds, precisions[:-1], label="Precision", marker='.')
plt.plot(thresholds, recalls[:-1], label="Recall", marker='.')
plt.xlabel("Threshold")
plt.ylabel("Score")
plt.title("Precision-Recall vs. Threshold - XGBoost")
plt.legend()
plt.grid(True)
plt.show()

# create thresholds between 0 and 1
thresholds = np.arange(0.1, 0.9, 0.01)

f1s = []
for thresh in thresholds:
    y_pred = (y_proba_xgb_rfe_sm >= thresh).astype(int)
    f1s.append(f1_score(y_test2, y_pred))

# find threshold with highest F1
best_thresh = thresholds[np.argmax(f1s)]
print(f"Best F1 threshold: {best_thresh:.2f}")
```

<span style="font-size:24px; font-weight:bold; color:black;">Hyperparameter Tuning</span>

```python
# Define the hyperparameter grid
param_dist = {
    'n_estimators': [100, 200, 300, 400],
    'max_depth': [3, 4, 5, 6, 7],
    'learning_rate': [0.01, 0.05, 0.1, 0.2],
    'subsample': [0.6, 0.8, 1.0],
    'colsample_bytree': [0.6, 0.8, 1.0],
    'gamma': [0, 0.1, 0.3, 0.5],
    'scale_pos_weight': [1, 2, 5, 10] 
}

# Initialize XGBoost model
xgb_tune = XGBClassifier(use_label_encoder = False, eval_metric = 'logloss', random_state = 42)

# Set up RandomizedSearchCV
rand_search = RandomizedSearchCV(
    estimator = xgb_tune,
    param_distributions = param_dist,
    n_iter = 50,  # number of combinations to try
    scoring = 'f1', 
    cv = 5,
    verbose = 1,
    random_state = 42,
    n_jobs =- 1
)

# Run search on RFE + SMOTE-balanced data
rand_search.fit(X_train_rfe_sm, y_train_rfe_sm)

# Print best params
print("Best Parameters Found:")
print(rand_search.best_params_)
```

```python
# Retrain final model with best params
best_xgb_model = rand_search.best_estimator_

# Predict again
y_pred_best = best_xgb_model.predict(X_test_rfe)
y_prob_best = best_xgb_model.predict_proba(X_test_rfe)[:, 1]

#  Store results
results_list.append({
    "model": "XGBoost (RFE + SMOTE) Tuned",
    "precision": precision_score(y_test, y_pred_best),
    "recall": recall_score(y_test, y_pred_best),
    "f1": f1_score(y_test, y_pred_best),
    "roc_auc": roc_auc_score(y_test, y_prob_best)
})
fitted_models["xgb_rfe_smote_tuned"] = best_xgb_model

#since the tuned model did worse than the default, I will stick with the original
```
